```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv')

df.head()
```

```python
types = pd.DataFrame(df.dtypes)
types.rename(columns = {0:'Type'}, inplace = True)
types
```

### Data Preprocessing


First, we need to convert categorical variables to binary variables. We will use pandas `get_dummies()` method for this.

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

For Rain Tommor we replace Yes with 1 and No with 0, without creating dummies.

```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

Now, we set our 'features' or x values and our Y or target variable.

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
df_sydney_processed['RainTomorrow']
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

## Linear Regression

```python
X_train, X_test, Y_train, Y_test = train_test_split( features, Y, test_size=0.2, random_state=10)
```

```python
LinearReg = LinearRegression()
LinearReg.fit(X_train, Y_train)
predictions = LinearReg.predict(X_train)
predictions
```

Let us evaluate our model.

```python
LinearRegression_MAE = metrics.mean_absolute_error(Y_train, predictions)
LinearRegression_MSE = metrics.mean_squared_error(Y_train, predictions)
LinearRegression_R2 = metrics.r2_score(Y_train, predictions)
```

```python
Rep = [['MAE', LinearRegression_MAE], ['MSE', LinearRegression_MSE], ['R2', LinearRegression_R2]]
report = pd.DataFrame(Rep, columns=['Evaluation of LR', 'Value'])
report
```

## KNN

First we create model, and then we evaluate it.

```python
KNN = KNeighborsClassifier(n_neighbors = 4).fit(X_train,Y_train)
predictions = KNN.predict(X_test)
KNN_Accuracy_Score = metrics.accuracy_score(Y_test, predictions)
KNN_JaccardIndex = metrics.jaccard_score(Y_test, predictions)
KNN_F1_Score = metrics.f1_score(Y_test, predictions)
```

```python
print('KNN_Accuracy_Score =', KNN_Accuracy_Score)
print('KNN_JaccardIndex =', KNN_JaccardIndex)
print('KNN_F1_Score =', KNN_F1_Score)
```

## Decision Tree

Same with decision tree model.

```python
Tree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
Tree.fit(X_train,Y_train)
predictions = Tree.predict(X_test)
Tree_Accuracy_Score = metrics.accuracy_score(Y_test, predictions)
Tree_JaccardIndex = metrics.jaccard_score(Y_test, predictions)
Tree_F1_Score = metrics.f1_score(Y_test, predictions)
```

```python
print('DT Accuracy_Score =', Tree_Accuracy_Score) 
print('DT Jaccard Index =', Tree_JaccardIndex)
print('DT F1 Score =',Tree_F1_Score)
```

## Logistic Regression

For Logistic Regression we need to change our train/test split.

```python
x_train, x_test, y_train, y_test = train_test_split( features, Y, test_size=0.2, random_state=1)
```

```python
LR = LogisticRegression(solver='liblinear')
LR.fit(X_train, Y_train)
predictions = LR.predict(X_test)
LR_Accuracy_Score = metrics.accuracy_score(Y_test, predictions)
LR_JaccardIndex = metrics.jaccard_score(Y_test, predictions)
LR_F1_Score = metrics.f1_score(Y_test, predictions)
LR_Log_Loss = metrics.log_loss(Y_test, predictions)
```

```python
print('LR Accuracy Score =', LR_Accuracy_Score)
print('LR JaccardIndex =', LR_JaccardIndex)
print('LR F1 Score =', LR_F1_Score)
print('LR Log Loss =', LR_Log_Loss)
```

## SVM

For SVM model we repeat everything again.

```python
SVM = svm.SVC()
SVM.fit(X_train, Y_train)
predictions = SVM.predict(X_test)
SVM_Accuracy_Score = metrics.accuracy_score(Y_test, predictions)
SVM_JaccardIndex = metrics.jaccard_score(Y_test, predictions)
SVM_F1_Score = metrics.f1_score(Y_test, predictions)
```

```python
print('SVM_Accuracy_Score =', SVM_Accuracy_Score)
print('SVM_JaccardIndex =', SVM_JaccardIndex)
print('SVM_F1_Score =', SVM_F1_Score)
```
